```python
%matplotlib inline
import pandas as pd
pd.__version__  # for the record
```

```python
df = pd.read_csv('time_series_covid19_confirmed_US.csv')
```

```python
df.head()
df.tail()
```

```python
df.describe()
```

```python
df.columns
```

```python
cases = df.groupby('Country_Region').sum().iloc[:,5:].T
cases.columns
cases.columns = ['Cases']
cases.columns
cases.head()
```

```python
cases.plot()
```

```python
tdf = cases['Cases'].shift(1)
tdf.head()
tdf.tail()
```

```python
cases['New Cases'] = cases['Cases'] - cases['Cases'].shift(1)
```

```python
cases.head()
cases.tail()
```

```python
cases['New Cases'].plot()
```

```python
df = pd.read_csv('time_series_covid19_deaths_US.csv')
df.head()
```

```python
deaths = df.groupby('Country_Region').sum().iloc[:,6:].T
deaths.columns
deaths.columns = ['Deaths']
deaths.columns
deaths.head()
deaths.tail()
```

```python
deaths.plot()
```

```python
deaths['New Deaths'] = deaths['Deaths'] - deaths['Deaths'].shift(1)
```

```python
deaths.tail()
```

```python
deaths['New Deaths'].plot()
```

```python
df = pd.DataFrame([cases['New Cases'],deaths['New Deaths']]).T
```

```python
df.head()
```

```python
df['Deaths per Cases'] = df['New Deaths'] / df['New Cases']
df.head()
df.tail()
```

```python
df['Deaths per Cases'].plot(figsize=(16,8),grid=True)
```
